```python
#Importando as Livrarias que iremos utilizar

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Importando o dataset

data = pd.read_csv('euro2024_players.csv')
```

```python
#Visualizando as primeiras linhas

data.head(5)
```

```python
#Checando a quantidade de linhas e colunas no Dataset

data.shape
```

```python
#Checando valores nulos nas colunas

data.isnull().sum()
```

```python
# Criando uma lista com os índices das linhas onde 'Foot' é nulo
lista_nulos = data[data['Foot'].isnull()].index.tolist()

# Exibindo a lista
print(lista_nulos)
```

```python
#Pesquisando as linhas nulas
linhas_especificas = data.loc[[390, 418, 472]]

# Exibindo as linhas
print(linhas_especificas)
```

```python
# Substituindo os valores da coluna 'Foot' nas linhas 390, 418 e 472 por 'right'
data.loc[[390, 418, 472], 'Foot'] = 'right'

# Verificando as alterações
print(data.loc[[390, 418, 472]])
```

```python
#Checando os tipos de colunas do dataset

data.info()
```

```python
#Análise Exploratoria

#Jogador com mais Gols na competição
most_goals_index = data['Goals'].idxmax()
player_most_goals = data.loc[most_goals_index]

print("Jogador com mais gols:")
print(player_most_goals)
```

```python
#Jogador com mais partidas
most_caps_index = data['Caps'].idxmax()
player_most_caps = data.loc[most_caps_index,['Name','Caps']]

print("Jogador com mais partidas:")
print(player_most_caps)
```

Assim, sem surpresas, Cristiano Ronaldo é o jogador com mais partidas na Euro 2024 e também o primeiro colocado em número de gols marcados.

```python
#Análise de Jogadores por pé favorito.
foot = data['Foot'].value_counts()
plt.figure(figsize=(10, 6))
foot.plot(kind='bar')
plt.title('Distribuição de jogador de acordo com o pé dominante')
plt.xlabel('Foot')
plt.ylabel('Number of Players')
plt.show()
```

A maioria dos jogadores é destro.

```python
# Número total de gols marcados por jogadores de um país

df_grouped = data.groupby('Country').sum().reset_index()
plt.figure(figsize=(10, 15))
goalplot=sns.barplot(x='Goals',y='Country',data = df_grouped)
plt.title('Número total de gols marcados por jogadores de um país')
plt.xlabel('Number of Goals')
plt.ylabel('Name of Country')
for p in goalplot.patches:
    goalplot.annotate(format(p.get_width(), '.1f'), 
                     (p.get_width(), p.get_y() + p.get_height() / 2.), 
                     ha = 'center', va = 'center', 
                     xytext = (20, 0), 
                     textcoords = 'offset points')

plt.show()
```

Os jogadores de Portugal, como coletivo, marcaram o maior número de gols pelo seu país. A Albânia ocupa o último lugar com 40

```python
#Valor total de mercado dos jogadores de um país

plt.figure(figsize=(10, 15))
goalplot2=sns.barplot(x='MarketValue',y='Country',data = df_grouped)
plt.title('Valor total de mercado dos jogadores de um país')
plt.xlabel('Total Market Value')
plt.ylabel('Name of Country')
for p in goalplot2.patches:
    goalplot2.annotate(format(p.get_width(), '.1f'), 
                     (p.get_width(), p.get_y() + p.get_height() / 2.), 
                     ha = 'center', va = 'center', 
                     xytext = (20, 0), 
                     textcoords = 'offset points')

plt.show()
```

Infelizmente, o conjunto de dados não especifica nenhuma moeda específica, por isso vamos chamá-lo apenas de unidades. Os jogadores ingleses têm o maior valor de mercado combinado, com um total superior a 1,5 mil milhões de unidades.

```python
#Jogador com maior valor de mercado
most_mv_index = data['MarketValue'].idxmax()
player_max_value = data.loc[most_caps_index,['Name','MarketValue']]

print("Jogador com maior valor de mercado:")
print(player_max_value)
```

```python
# Criando uma lista de Valor total de mercado dos jogadores de um clube
data_group = data.groupby('Club').sum().reset_index()
top_10_clubs = data_group.sort_values(by='MarketValue', ascending=False).head(10)
```

```python
#Top 10 valor total de mercado dos jogadores de um clube
plt.figure(figsize=(10, 15))
sns.barplot(x='MarketValue',y='Club',data = top_10_clubs)
plt.title('Valor total de mercado dos jogadores de um clube (Top 10)')
plt.xlabel('Total Market Value')
plt.ylabel('Name of Clubs')
plt.show()
```

Vemos que o Manchester City é onde se encontra o maior valor de jogadores nessa competição.
